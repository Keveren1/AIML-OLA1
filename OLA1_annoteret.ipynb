## Data Exploration and Cleaning

```python
import pandas as pd

# Indl칝s datas칝ttet
df = pd.read_csv("events.csv")

# F칮rste kig p친 datastrukturen
df.head()
```

```python
# Vis info om kolonner og datatyper
df.info()

# Statistisk beskrivelse af numeriske kolonner
df.describe()
```

游닇 **Kommentar:**  
- Datas칝ttet indeholder over 940.000 h칝ndelser.  
- `event_type`, `side`, `event_team`, `player` er eksempler p친 kategoriske data.  
- `time`, `sort_order`, `is_goal` er numeriske v칝rdier.  
- Nogle kolonner har mange manglende v칝rdier, som vi behandler n칝ste trin.

```python
# Tjek for manglende v칝rdier
df.isnull().sum()
```

**Observationer og plan:**  
F칮lgende kolonner har >50% manglende v칝rdier og droppes:  
- `event_type2`, `player_in`, `player_out`, `shot_place`, `shot_outcome`, `bodypart`, `situation`, `location`  

Vi beholder `player` samt `player2`, og udfylder dem med `"Unknown"`.


```python
# Drop kolonner med for mange manglende v칝rdier
df_clean = df.drop(columns=[
    'event_type2', 'player_in', 'player_out',
    'shot_place', 'shot_outcome', 'bodypart', 'situation', 'location'
])

# Udfyld player med 'Unknown'
df_clean['player'] = df_clean['player'].fillna('Unknown')

# Vi g칮r det samme for player2
df_clean['player2'] = df_clean['player2'].fillna('Unknown')

# Bekr칝ft at alle manglende v칝rdier er h친ndteret
df_clean.isnull().sum()
```

```python
df_clean.info()
df_clean.describe()
df_clean.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
sns.histplot(df_clean['time'], bins=90)
plt.title("Fordeling af h칝ndelser over kampens varighed")
plt.xlabel("Tid i minutter")
plt.ylabel("Antal h칝ndelser")
plt.show()
```

```python
sns.countplot(data=df_clean, x='side')
plt.title("H칝ndelser fordelt p친 side af banen")
plt.show()
```

## Task 2: Feature Engineering & Descriptive Statistics

```python
# Eksempel: Del kampen op i 3 perioder
bins = [0, 30, 60, 100]
labels = ['Early', 'Mid', 'Late']
df_clean['match_period'] = pd.cut(df_clean['time'], bins=bins, labels=labels, right=False)

# Tjek fordelingen
df_clean['match_period'].value_counts()
```

```python
df_encoded = pd.get_dummies(df_clean, columns=['side'], prefix='side')

# Statistiske m친l for 'time'
mean_time = df_encoded['time'].mean()
median_time = df_encoded['time'].median()
std_time = df_encoded['time'].std()

print(f"Gennemsnit: {mean_time:.2f}, Median: {median_time}, Std: {std_time:.2f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot for 'time' pr. kampperiode
sns.boxplot(data=df_encoded, x='match_period', y='time')
plt.title("Fordeling af tid pr. periode")
plt.show()

# Barplot over antal h칝ndelser pr. periode
sns.countplot(data=df_encoded, x='match_period')
plt.title("Antal h칝ndelser i hver kampperiode")
plt.show()
```
